```python
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:90% !important; }</style>"))

from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)

from jupyterthemes import jtplot
jtplot.style()

from talib.abstract import *
import pandas as pd
import os
import numpy as np
import plotly.graph_objs as go
from datetime import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
directory = 'SET_EOD_2010_2016'
files = os.listdir(directory)
files.sort()

frames = []
for file in files:
    df = pd.read_csv(os.path.join(directory, file))
    df['<DTYYYYMMDD>'] = pd.to_datetime(df['<DTYYYYMMDD>'], format='%Y%m%d')
    frames.append(df)
```

```python
merged = pd.concat(frames)
merged = pd.DataFrame.reset_index(merged)
merged = merged.drop('index', axis=1)
```

```python
merged.head()
merged.tail()
```

```python
merged.to_csv('merged.csv', index=False)
```

### Merged "SET_EOD_2010_2016" to single CSV file

```python
data = pd.read_csv('merged_2010_2016.csv')
data['Date'] = pd.to_datetime(data['Date'], format='%Y-%m-%d')
```

```python
data.head()
data.info()
```

```python
stock = data.loc[(data['Date'].dt.year==2016) & (data['Ticker']=='ADVANC')]
stock = stock.set_index('Date')
stock.drop('Volume', axis=1).plot()
```

## Time Series Plot with datetime Objects

```python
data = [go.Scatter(x=stock.index, y=stock.Close)]

iplot(data)
```

## Time Series With Rangeslider

```python
trace_high = go.Scatter(
    x=stock.index,
    y=stock.High,
    name = "ADVANC High",
    line = dict(color = '#17BECF'),
    opacity = 0.8)

trace_low = go.Scatter(
    x=stock.index,
    y=stock.Low,
    name = "ADVANC Low",
    line = dict(color = '#7F7F7F'),
    opacity = 0.8)

data = [trace_high,trace_low]

layout = dict(
    title='Time Series with Rangeslider',
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1,
                     label='1m',
                     step='month',
                     stepmode='backward'),
                dict(count=6,
                     label='6m',
                     step='month',
                     stepmode='backward'),
                dict(step='all')
            ])
        ),
        rangeslider=dict(),
        type='date'
    )
)

fig = dict(data=data, layout=layout)
iplot(fig, filename = "Time Series with Rangeslider")
```

## Candlestick Charts

```python
trace = go.Candlestick(x=stock.index,
                       open=stock.Open,
                       high=stock.High,
                       low=stock.Low,
                       close=stock.Close)

iplot([trace], filename='simple_candlestick')
```

```python
sma35 = SMA(stock, timeperiod=35, price='Close')
sma15 = SMA(stock, timeperiod=15, price='Close')
sma15.tail()
```

```python
trace_vol = go.Bar(
            x=stock.index,
            y=stock.Volume
    )

iplot([trace_vol])
```

```python
trace_SMA15 = go.Scatter(
    x=sma15.index,
    y=sma15.values,
    name = "ADVANC SMA15",
    line = dict(color = '#17BECF'),
    opacity = 0.8)

trace_SMA35 = go.Scatter(
    x=sma35.index,
    y=sma35.values,
    name = "ADVANC SMA35",
    line = dict(color = '#7F7F7F'),
    opacity = 0.8)

data = [trace_SMA15, trace_SMA35, trace]

layout = dict(
    title='Time Series with Rangeslider',
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1,
                     label='1m',
                     step='month',
                     stepmode='backward'),
                dict(count=6,
                     label='6m',
                     step='month',
                     stepmode='backward'),
                dict(step='all')
            ])
        ),
        rangeslider=dict(),
        type='date'
    )
)

fig = dict(data=data, layout=layout)
iplot(fig, filename = "Time Series with Rangeslider")
```
